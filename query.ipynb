```python
import requests
from pprint import pprint
```

```python
BASE_URL = "http//localhost:8000/api/productos/"
```

```python
response = requests.get(BASE_URL)
print('Status:', response)
print('Status:', response.status_code)
try:
    print(response.json())
except Exception as e:
    print("No se pudo decodificar la respuesta como JSON. Respuesta cruda:")
    print(response.text)
```

```python
nuevo_producto = {
    "nombre": "Producto desde notebook",
    "descripcion": "Creado via Jupyter",
    "precio": 19.99,
}
response = requests.post(BASE_URL, data=nuevo_producto)
print('Status:', response.status_code)
try:
    print(response.json())
except Exception as e:
    print("No se pudo decodificar como JSON. Respuesta cruda:")
    print(response.text)
```
