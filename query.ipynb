```python
import requests
from pprint import pprint
```

```python
# URLs de la API
BASE_URL = "http://localhost:8000/api/usuarios/"
TARGET_URL = "http://localhost:8000/api/aocardo-target/"

print('BASE_URL =', BASE_URL)
print('TARGET_URL =', TARGET_URL)
```

```python
# Establecer una palabra objetivo (ejemplo: 'perro')
payload = {'palabra': 'avestruz'}  # Cambia aquÃ­ la palabra del juego
resp = requests.post(TARGET_URL, json=payload)
print('Set target status:', resp.status_code)
try:
    pprint(resp.json())
except Exception:
    print(resp.text)
```

```python
response = requests.get(BASE_URL)
print('Status:', response.status_code)
try:
    print(response.json())
except Exception as e:
    print("No se pudo decodificar la respuesta JSON. Respuesta cruda:")
    print(response.text)
```
